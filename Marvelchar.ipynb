```
import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```
url = "https://marvel-thor.surge.sh/"
page = urlopen(url)
```

```
#soup = BeautifulSoup(page, 'html.parser')
soup = BeautifulSoup(page, 'lxml')
#print(soup)
```

```
name=[]
characterstics=[]
Identity=[] 
Align=[]
Eye=[]
Hair=[]
Sex=[]
Alive=[]
Appearances=[]
First_Appearance=[]
Year=[]
i=0
j=0
whole_data = soup.find_all('div', attrs={'class': "col-lg-8 mx-auto"})
#name_box = soup.find_all('h2')
#print(len(whole_data))
for eachiter in whole_data:
  j=j+1
  x=eachiter.find_all('p',attrs={'class':'h2'})
  cleanx=BeautifulSoup(str(x), "lxml").get_text()
  cleanx=cleanx.strip('[]')
  #print(len(cleanx))
  if len(cleanx)==0:
    i=i+1
    x=eachiter.find_all('h2')
    cleanx=BeautifulSoup(str(x), "lxml").get_text()
    cleanx=cleanx.strip('[]')
    #print(i,cleanx)
    #print(i)
    #print(cleanx)
  name.append(cleanx)
  #print(name)
  
  #y=eachiter.find_all('p',attrs={'class':'h4'})
  y=eachiter.find_all('a', href=True)
  cleany=BeautifulSoup(str(y), "lxml").get_text()
  cleany=cleany.strip('[]')
  cleany=cleany.split(',')
  characterstics.append(cleany)
  #if j==2:
  #  print(cleany)
  #print(y['href'])
  #if strip(cleany)="Identity":
  #  i=i+1
  #Identity.append(cleany)
  
#print(len(name))
#print(name)
print(type(characterstics))
print(len(characterstics))
```
